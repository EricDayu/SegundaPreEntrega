**ABSTRACTO**

La NBA (National Basketball Association) es una liga privada de Basketball profesional creada en Estados Unidos. Está conformada por 30 equipos o franquicias, estando 29 en Estados Unidos y 1 en Canadá. Se ha visto que hay dos grandes conjuntos de factores que determinan el éxito a largo plazo de una franquicia en la NBA. Existen factores coyunturales, que refieren al mercado donde se encuentra el cuadro, y otros relacionados a la capacidad del equipo de maximizar las oportunidades y por ende su valor. Los coyunturales son incontrolables, generalmente se asocian a la ciudad del equipo. Dentro de estos factores se incluye, a) tamaño del mercado, que tiene impacto en la cantidad de dinero que puede ingresar mediante sponsors, ventas de entradas, etc.; b) tamaño de la ciudad, siendo favoritas las grandes ciudades; c) clima de dicha ciudad; entre otros. Sin embargo, el manejo de un equipo puede tener distintos resultados, llevándolo al éxito o fracaso según las decisiones tomadas. Un gran punto para controlar es la elección de los jugadores, que deben ser parte de la NBA. Los equipos pueden contratar en la agencia libre, es decir, jugadores que son parte de otro equipo y a partir del nuevo contrato cambian de equipo. El acceso a la NBA por primera vez se realiza mediante un proceso de selección llamado “Draft”, donde los jugadores pasan a ser parte de ciertos equipos. En esta instancia, todos los potenciales nuevos jugadores, que generalmente provienen de universidades, se inscriben en la elección del Draft y luego los equipos van eligiendo a los jugadores en un sistema que promueve la paridad. Se realiza un sorteo para definir el orden de elección de los jugadores favoreciendo a los equipos que salieron en las peores posiciones en el campeonato. No obstante, lo que realmente importa es qué tan buenos sean los equipos identificando cuáles serán los mejores jugadores de la NBA y en base a eso hacer la elección. De nada servirá elegir antes si no logran identificar los futuros mejores jugadores. A su vez, el rendimiento y desarrollo de un jugador también se verán influenciados por la preparación que provea cada equipo.

Nosotros somos una empresa de asesoramiento deportivo que busca asesorar a equipos y representantes de jugadores mediante la Ciencia de Datos. En cuanto al servicio hacia los equipos, buscamos predecir el rendimiento futuro de los jugadores con el fin de recomendar o no su contratación. Por ejemplo, se estudiará de cuáles universidades han salido los mejores jugadores en el pasado. Por otro lado, los representantes buscarán que sus jugadores sean contratados por cuadros que vayan a tener éxito. Para ello, se analizará el rendimiento de los jugadores de los distintos equipos. Se hipotetiza que el rendimiento se puede predecir en base a ciertas estadísticas de los jugadores. Por lo tanto, se buscará la correlación de distintas estadísticas con sus rendimientos, para luego en base a los datos disponibles, poder predecir qué equipos y universidades preparan mejor a sus jugadores.

**Algoritmo y Validación de Modelos**

El algoritmo busca predecir el valor de VORP de un determinado jugador utilizando las otras variables. Se buscará generar un modelo y luego validarlo con los datos del dataset.

```python
#Definimos X e y, y sacamos las features no numéricas ni relevantes como id
X = df_nba2.drop(['value_over_replacement','id','team','player','college'], axis=1)
y = df_nba2['value_over_replacement']

X.shape, y.shape
```

```python
import joblib
import sys 
sys.modules['sklearn.externals.joblib']=joblib
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.linear_model import LinearRegression
```

```python
#Evaluamos cuáles features son las 6 más relevantes para el vorp
lreg = LinearRegression()
sfs1 = sfs(lreg, k_features=6, forward=True, verbose=2, scoring='neg_mean_squared_error')
```

```python
sfs1 = sfs1.fit(X, y)
```

```python
feat_names = list(sfs1.k_feature_names_)
print ("Las features más relevantes son:", feat_names)
```

Según el análisis, las features más relevantes son las mencionadas anteriormente. Nos llama la atención no encontrar en el listado a las que más correlación mostraron en el heatmap y encontrar a otras con mala correlación. Podría ser que en su conjunto correlacionen mejor que uno a uno. 

Se generará un modelo utilizando estas 6 variables y otro utilizando las 6 variables de mayor correlación según el heatmap y se compararán. 

```python
#Modelo con variables de algoritmo
#creamos un nuevo data frame con las 6 variables +el VORP y eliminamos el registro para el que falta el dato de una de las variables para que después pueda funcionar el algoritmo
new_nba2 = df_nba2[['year', 'rank', 'overall_pick', 'years_active', 'games', 'field_goal_percentage','value_over_replacement']]
df_nba2_drop = new_nba2.dropna(subset=['field_goal_percentage'])

df_nba2_drop
```

```python
#Volvemos a definir X e y porque sacamos valores nulos para que funcione el algoritmo
X2 = df_nba2_drop.drop(['value_over_replacement'], axis=1)
y2 = df_nba2_drop['value_over_replacement']


X.shape, y.shape
```

**Entrenamiento del modelo**

```python
#Entrenando el modelo con el 70% de los datos

from sklearn.model_selection import train_test_split

X2_train,X2_test,y2_train,y2_test = train_test_split(X2, y2 ,test_size=0.3,random_state=2)

# Ajustar el modelo con X2_train y y2_train

lreg.fit(X2_train,y2_train)

# PRedecir con X2_test

y2_pred = lreg.predict(X2_test)
```

**Evaluación del modelo**

```python
from sklearn.metrics import r2_score

r2 = r2_score(y2_test,y2_pred)
print(r2)
```

```python
figypred, ax = plt.subplots(figsize=(8, 8))  
ax.scatter(y2_test, y2_pred)
plt.title('Evaluación del vorp predicho')
plt.ylabel('Vorp predicho')
plt.xlabel('Vorp real')

m, b = np.polyfit(y2_test, y2_pred, 1)

#add linear regression line to scatterplot 
plt.plot(y2_test, m*y2_test+b)
```

El modelo que se generó utilizando las variables que el algoritmo  marcó no es bueno, se ve reflejado en el r2 de 0.44 y en la gráfica. Se tendrá que ajustar el modelo para predecir el VORP.

Probamos un modelo con las variables que correlacionan mejor según el heatmap.

```python
#Modelo con variables de heatmap
#creamos un nuevo data frame con las 6 variables 
hm_nba2 = df_nba2[['win_shares','points','assists','minutes_played','total_rebounds','games','value_over_replacement']]
hm_nba2
```

```python
#Definimos X e y para este modelo
X3 = hm_nba2.drop(['value_over_replacement'], axis=1)
y3 = hm_nba2['value_over_replacement']


X.shape, y.shape
```

**Entrenamiento del nuevo modelo**

```python
#Entrenando el modelo con el 70% de los datos

from sklearn.model_selection import train_test_split

X3_train,X3_test,y3_train,y3_test = train_test_split(X3, y3 ,test_size=0.3,random_state=2)

# Ajustar el modelo con X3_train y y3_train

lreg.fit(X3_train,y3_train)

# PRedecir con X3_test

y3_pred = lreg.predict(X3_test)
```

**Evaluación del nuevo modelo**

```python
r2 = r2_score(y3_test,y3_pred)
print(r2)

figypred, ax = plt.subplots(figsize=(8, 8))  
ax.scatter(y3_test, y3_pred)
plt.title('Evaluación del vorp predicho con variables del heatmap')
plt.ylabel('Vorp predicho')
plt.xlabel('Vorp real')

m, b = np.polyfit(y3_test, y3_pred, 1)

#add linear regression line to scatterplot 
plt.plot(y3_test, m*y3_test+b)
```

El modelo que se genera utilizando las variables que mejor correlacionan según el heatmap tiene un r2 de 0.92, por lo que consideramos que es un buen modelo. Esto también se ve reflejado en la gráfica que compara el vorp predicho y el real. Se podría seguir mejorando el modelo buscando una mejor correlación pero se dejará para el futuro. Con estas variables consideramos que ya hay un modelo aceptable que puede ser utilizado para el objetivo inicial de predecir el VORP de un jugador para poder predecir si será bueno o no.
